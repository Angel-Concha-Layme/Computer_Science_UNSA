# Ecuaci√≥n 4

```python
#resolver oscilaciones forzadas de forma generalizada 
# Cociente F/m=1
#Frecuencia angular propia, œâ0=100 rad/s 
# Frecuencia de la fuerza oscilante, œâf=120 rad/s

import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt


w0 = 100; #frecuencia angular propia
g = 7; #rozamiento, gamma,
wf = 100; #frecuencia de la fuerza oscilante (cambiar este valor)
F = 1; #F0/m amplitud de la fuerza oscilante
#condiciones iniciales
x0 = [0,-0.08]; #posici√≥n inicial, x0, velocidad inicial, v0
tf = 0.3*np.pi; #tiempo final
f = lambda t,x: [x[2],-2*g*x[2]-w0*w0*x[1]+F*np.cos(wf*t)]
tspan = [0, tf]


#plot 
t = np.linspace(0, tf, 1000)

#plot python
def graficar():
    plt.plot(t, solucion(t, x0), color = 'red')
    plt.title('Onda')
    plt.xlabel('Tiempo')
    plt.ylabel('Amplitud')
    plt.grid(True)
    plt.show()

#solucion
def solucion(t, x0):
    sol = integrate.solve_ivp(f, tspan, x0, t_eval=t)
    return sol.y[1]

print ('La amplitud es: ', solucion(t, x0))

graficar()









```

# Ecuaci√≥n 6

```python
Q0 = float(input('Ingrese el valor de Q0: ')) # Q0 es la carga inicial de la bateria
œâ = float(input('Ingrese el valor de œâ: ')) # œâ es la frecuencia angular
œÜ = float(input('Ingrese el valor de œÜ: ')) # œÜ es la fase inicial de la onda
t = np.linspace(0, 2*np.pi, 1000) # t es el tiempo en segundos

def carga(Q0, œâ, œÜ):
    return Q0*np.sin(œâ*t + œÜ)

def graficar():
    plt.plot(t, carga(Q0, œâ, œÜ), color = 'red')
    plt.title('Carga')
    plt.xlabel('Tiempo')
    plt.ylabel('Carga')
    plt.grid(True)
    plt.show()

print ('La carga es: ', carga(Q0, œâ, œÜ))
graficar()
```
