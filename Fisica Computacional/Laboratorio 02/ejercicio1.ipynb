```python
masa1= float (input ("Masa 1: " ))
masa2= float (input ("Masa 2: " ))

gravedad = 9.8

peso1 = masa1 * gravedad
peso2 = masa2 * gravedad

if masa1 == masa2:
    aceleracion = 0

elif masa1 > masa2:
    aceleracion = ((masa1-masa2)/(masa1+masa2))*gravedad
    tension1 = peso1 - masa1 * aceleracion
    tension2 = peso2 + masa2 * aceleracion
elif masa1 < masa2:
    aceleracion = ((masa2-masa1)/(masa1+masa2))*gravedad
    tension1 = peso1 + masa1 * aceleracion
    tension2 = peso2 - masa2 * aceleracion

print ( "Aceleracion: " , aceleracion)

print ( "Tension 1: " , tension1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

masa = float (input ( "Masa: " ))
distancia = float (input ( "Distancia: " ))
tiempo = float (input ( "Tiempo: " ))
v_inicial = float (input ( "Velocidad inicial: " ))
v_final = float (input ( "Velocidad final: " ))

aceleracion = (v_final - v_inicial) / tiempo
fuerza = masa * aceleracion

print("Fuerza:",fuerza)

t = np.linspace(0,tiempo,100)
x = v_inicial * t + 0.5 * aceleracion * t**2

plt.plot(t,x)
plt.xlabel( "Tiempo" )
plt.ylabel( "Distancia" )


```
